```python
import pandas as pd
```

```python
student_data = pd.read_excel('students/students_info.xlsx')
results = pd.read_html('students/results_ejudge.html')[0]
```

```python
student_data.head()
```

```python
results.head()
```

```python
df = results.merge(student_data, left_on='User', right_on='login')
df.head()
```

```python
solved_by_group = df[['group_faculty', 'Solved']].groupby(['group_faculty']).mean()
solved_by_group.plot(kind='bar', title='Mean solved', legend=False)
```

```python
solved_by_group_out = df[['group_out', 'Solved']].groupby(['group_out']).mean()
solved_by_group_out.plot(kind='bar', title='Mean solved', legend=False)
```

```python
top_users = df[(df.G>=10) | (df.H>=10)]
top_users
```

```python
top_users.groupby(['group_faculty']).size().plot(kind='bar', title='Group for top users')
```

```python
top_users.groupby(['group_out']).size().plot(kind='bar', title='Group for top users')
```
