```python
from sympy import *
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
A = []
b = []
def initSystem(file_name):
    global A, b
    with open(f"{file_name}") as f:
        lines = f.readlines()

    lines = [list(map(float, line.strip().split())) for line in lines]
    N = int(lines[0][0])
    A = lines[1:N+1]
    b = lines[N+1]
```

```python
def solveSystem():
    global A, b
    x = linalg.solve(A, b)
    plt.bar(range(len(x)), x)
    plt.gca().grid(True)
```

```python
initSystem('small.txt')
solveSystem()
```

```python
initSystem('large.txt')
solveSystem()
```
