```python
from PIL import Image
import numpy as np
```

```python
img = Image.open('lunar_images/lunar03_raw.jpg')
data = np.array(img)
```

```python
min = data.min()
max = data.max()
```

```python
min
```

```python
max
```

```python
data = data - min
data = data * (255/(max-min))
```

```python
res_img = Image.fromarray(data)
res_img.convert("L").save('lunar_images/lunar03_result.jpg')
```
