```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=3
```

```python
data = np.loadtxt('signals/signal0{}.dat'.format(n), dtype=float)
```

```python
newData = []
def lowerIndex(i):
    return (i-9) if (i-9)>=0 else 0
    
for i in range(len(data)):
    newData.append(np.mean(data[lowerIndex(i):i+1]))
```

```python
plt.plot(data)
plt.ylim(0, 35)
plt.xlim(-1, 101)
plt.show()
```

```python
plt.plot(newData)
plt.ylim(0, 35)
plt.xlim(-1, 101)
plt.show()
```
